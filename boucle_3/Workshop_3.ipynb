````python
import numpy as np
import plotly.graph_objects as go
x1_max = 5
x1 = np.linspace(0, x1_max) # on genere l'intervalle des x
x2 = 4 - 2*x1 # on genere l'intervalle des y
x2_max = np.amax(x2)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x1, y=x2, mode='lines', name='x2', line=dict(color='royalblue')))
fig.add_trace(go.Scatter(
    x=x1,
    y=np.full_like(x1, x2_max),
    mode='lines',
    line=dict(color='lightgrey', width=0),
    fill='tonexty',
    fillcolor='lightgrey',
    showlegend=False,
    hoverinfo='skip',
))
# on dessine entre 0,0 et les extremas des fonctions
fig.update_xaxes(range=[0, x1_max])
fig.update_yaxes(range=[0, x2_max])
# on affiche
fig.show()
````

On a donc tout ce qu'il nous faut pour afficher les inéquations de nos contraintes. Traçons les droites en rouge, et les zones d'inégalités en gris.



````python
import numpy as np
import plotly.graph_objects as go
xmax = 20
x1 = np.arange(0, xmax)
# Contraintes
y1 = 8 - 2*x1
y2 = (7 - x1)/2
y3 = np.full_like(x1, 3)  # ligne horizontale y = 3
# Valeur maximale utile du graphe
ymax = max(np.max(y1), np.max(y2), np.max(y3))
# Tracé des contraintes
fig = go.Figure()
fig.add_trace(go.Scatter(x=x1, y=y1, name='2x + y ? 8', line=dict(color='red')))
fig.add_trace(go.Scatter(x=x1, y=y2, name='x + 2y ? 7', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=x1, y=y3, name='y ? 3', line=dict(color='green')))
# Remplissage de la zone faisable (en dessous des trois contraintes)
y_feasible = np.minimum(np.minimum(y1, y2), y3)
fig.add_trace(go.Scatter(
    x=np.concatenate([x1, x1[::-1]]),
    y=np.concatenate([y_feasible, np.zeros_like(x1)]),
    fill='toself',
    fillcolor='lightgrey',
    line=dict(color='rgba(0,0,0,0)'),
    showlegend=False,
    hoverinfo='skip',
))
# Axes et limites
fig.update_xaxes(range=[0, xmax], title='x (kg de yaourt A)')
fig.update_yaxes(range=[0, ymax], title='y (kg de yaourt B)')
fig.update_layout(
    title='Zone faisable ? Problème de fabrication de yaourts',
    width=1200,
    height=700,
    template='plotly_white',
)
fig.show()
````

Nous allons pouvoir utiliser cette propriété pour définir la solution optimale. Il nous faut donc maintenant représenter la fonction économique. Sa forme générale est $ax +by = p$. Cette fonction est la famille des droites correspondant à l’ensemble–solution avec une pente $\dfrac{-a}{b}$.

Un [Widget](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html) intégré [automatiquement](https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html) à notre fonction de calcul nous permettra de faire varier facilement ce paramètre et de déterminer les valeurs admissibles de $X_1$ et $X_2$ maximisant la fonction économique. Une fois que vous avez localisé globalement la position de la solution, vous pouvez modifier l'intervalle `x` pour zoomer sur cette partie et obtenir un résultat plus précis.

````python
from ipywidgets import interact
import ipywidgets as widgets
import plotly.graph_objects as go
# cette fonction est appel?e ? chaque changement d'?tat du slider.
def trace(n):
    # on reprend le code precedent
    xmax = 20
    x1 = np.arange(0, xmax) # on genere l'intervalle des x
    # Contraintes
    y1 = 8 - 2*x1
    y2 = (7 - x1)/2
    y3 = np.full_like(x1, 3)  # ligne horizontale y = 3
    # Valeur maximale utile du graphe
    ymax = max(np.max(y1), np.max(y2), np.max(y3))
    # Trac? des contraintes
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x1, y=y1, name='2x + y ? 8', line=dict(color='red')))
    fig.add_trace(go.Scatter(x=x1, y=y2, name='x + 2y ? 7', line=dict(color='blue')))
    fig.add_trace(go.Scatter(x=x1, y=y3, name='y ? 3', line=dict(color='green')))
    # Remplissage de la zone faisable (en dessous des trois contraintes)
    y_feasible = np.minimum(np.minimum(y1, y2), y3)
    fig.add_trace(go.Scatter(
        x=np.concatenate([x1, x1[::-1]]),
        y=np.concatenate([y_feasible, np.zeros_like(x1)]),
        fill='toself',
        fillcolor='lightgrey',
        line=dict(color='rgba(0,0,0,0)'),
        showlegend=False,
        hoverinfo='skip',
    ))
    # Axes et limites
    fig.update_xaxes(range=[0, xmax], title='x (kg de yaourt A)')
    fig.update_yaxes(range=[0, ymax], title='y (kg de yaourt B)')
    pente = -4/5
    yeco = n + pente*x1
    fig.add_trace(go.Scatter(x=x1, y=yeco, name='Fonction économique', line=dict(color='black', dash='dash')))
    fig.update_layout(
        title='Zone faisable ? Problème de fabrication de yaourts',
        width=1200,
        height=700,
        template='plotly_white',
    )
    fig.show()
# le slider qui permet de faire varier la valeur passee en parametre de la fonction trace
interact(trace, n=widgets.FloatSlider(min=-1, max=20.0, step=1, value=-150.0));
````

# 2. Programme linéaire en nombre réels à $n$ dimensions

Arcelor-Mittal, géant mondial de l'acier, a ouvert une nouvelle usine en Grèce. Il fabrique 9 alliages à base de 3 matières premières&nbsp;: le plomb, le zinc et l'étain.
Voici un tableau avec les pourcentages de chaque matière dans l'alliage et son coût par unité&nbsp;:

| Alliage       | 1   | 2   | 3   | 4   | 5   | 6   | 7   | 8   | 9   |
|---------------|-----|-----|-----|-----|-----|-----|-----|-----|-----|
|**% plomb**    | 20  | 50  | 30  | 30  | 30  | 60  | 40  | 10  | 10  |
|**% zinc**     | 30  | 40  | 20  | 40  | 30  | 30  | 50  | 30  | 10  |
|**% étain**    | 50  | 10  | 50  | 30  | 40  | 10  | 10  | 60  | 80  |
|**Coût unit.** | 7.3 | 6.9 | 7.3 | 7.5 | 7.6 | 6.0 | 5.8 | 4.3 | 4.1 |

Le département de R&D recherche un « nouvel alliage » à prix minimum qui serait un mélange des 9 autres avec les contraintes d'avoir au maximum 30% de plomb, 30% de zinc et 40% d'étain.

## 2.1 Modélisation du problème

Commençons par modéliser le problème sous la forme d'un programme linéaire. Appliquez la méthode de l'exercice précédent.<br />

1. Identifiez les variables de décision du programme linéaire.

Les variables de décision sont les quantités de chaque alliage à mélanger pour obtenir le nouvel alliage. On peut les nommer $X_1$, $X_2$, ..., $X_9$ pour représenter les quantités des alliages 1 à 9 respectivement.

2. Identifiez les contraintes du programme linéaire.

Les contraintes sont basées sur les pourcentages de plomb, de zinc et d'étain dans le nouvel alliage. On doit s'assurer que le mélange final respecte les limites maximales de chaque matière première. Les contraintes peuvent être formulées comme suit :

- Pour le plomb : $0.20X_1 + 0.50X_2 + 0.30X_3 + 0.30X_4 + 0.30X_5 + 0.60X_6 + 0.40X_7 + 0.10X_8 + 0.10X_9 \leq 0.30$
- Pour le zinc : $0.30X_1 + 0.40X_2 + 0.20X_3 + 0.40X_4 + 0.30X_5 + 0.30X_6 + 0.50X_7 + 0.30X_8 + 0.10X_9 \leq 0.30$
- Pour l'étain : $0.50X_1 + 0.10X_2 + 0.50X_3 + 0.30X_4 + 0.40X_5 + 0.10X_6 + 0.10X_7 + 0.60X_8 + 0.80X_9 \leq 0.40$
- La somme des quantités de chaque alliage doit être égale à 1 : $\sum_{i=1}^9 X_i = 1$
- De plus, on doit s'assurer que les quantités de chaque alliage sont non négatives : $X_i \geq 0$, $\forall$ $i = 1, 2, ..., 9$

3. Exprimez la fonction économique du problème.

La fonction économique à minimiser est le coût total du mélange, qui peut être exprimé comme suit :
$Z = 7.3X_1 + 6.9X_2 + 7.3X_3 + 7.5X_4 + 7.6X_5 + 6.0X_6 + 5.8X_7 + 4.3X_8 + 4.1X_9$

4. Réunissez les réponses du dessus pour exprimer le programme linéaire complet optimisant la composition des alliages de l'usine.

Le programme linéaire complet peut être formulé comme suit :

Minimiser $Z = 7.3X_1 + 6.9X_2 + 7.3X_3 + 7.5X_4 + 7.6X_5 + 6.0X_6 + 5.8X_7 + 4.3X_8 + 4.1X_9$

s.c. (= sous contraintes)

$\begin{cases}
0.20X_1 + 0.50X_2 + 0.30X_3 + 0.30X_4 + 0.30X_5 + 0.60X_6 + 0.40X_7 + 0.10X_8 + 0.10X_9 \leq 0.30 \\
0.30X_1 + 0.40X_2 + 0.20X_3 + 0.40X_4 + 0.30X_5 + 0.30X_6 + 0.50X_7 + 0.30X_8 + 0.10X_9 \leq 0.30 \\
0.50X_1 + 0.10X_2 + 0.50X_3 + 0.30X_4 + 0.40X_5 + 0.10X_6 + 0.10X_7 + 0.60X_8 + 0.80X_9 \leq 0.40 \\
\sum_{i=1}^9 X_i = 1 \\
X_i \geq 0, \forall i = 1, 2, ..., 9
\end{cases}$


## 2.2 Résolution
Nous sommes en présence d'un problème à 9 variables. La méthode de résolution graphique ne peut pas être appliquée ici, nous allons nous tourner vers le solveur [PuLP](https://pythonhosted.org/PuLP/) qui calculera lui-même la solution, en exploitant la propriété qu'on a identifiée dans l'exercice précédent.


On va s'inspirer d'[un exemple basique](https://www.coin-or.org/PuLP/CaseStudies/a_blending_problem.html) pour implémenter et résoudre le modèle&nbsp;:



````python
from pulp import *
import numpy as np

# variables du problème
X1 = LpVariable("X_1", 0, None) # Nom de la variable, borne inférieure, borne supérieure
X2 = LpVariable("X_2", 0, None)
X3 = LpVariable("X_3", 0, None)
X4 = LpVariable("X_4", 0, None)
X5 = LpVariable("X_5", 0, None)
X6 = LpVariable("X_6", 0, None)
X7 = LpVariable("X_7", 0, None)
X8 = LpVariable("X_8", 0, None)
X9 = LpVariable("X_9", 0, None)


# problème
prob = LpProblem("alliage", LpMinimize) # Nom du problème, fonction objectif

# objectif
prob +=  7.3*X1 + 6.9*X2 + 7.3*X3 + 7.5*X4 + 7.6*X5 + 6.0*X6 + 5.8*X7 + 4.3*X8 + 4.1*X9

# contraintes
prob +=  0.20*X1 + 0.50*X2 + 0.30*X3 + 0.30*X4 + 0.30*X5 + 0.60*X6 + 0.40*X7 + 0.10*X8 + 0.10*X9 <= 0.30
prob +=  0.30*X1 + 0.40*X2 + 0.20*X3 + 0.40*X4 + 0.30*X5 + 0.30*X6 + 0.50*X7 + 0.30*X8 + 0.10*X9 <= 0.30
prob +=  0.50*X1 + 0.10*X2 + 0.50*X3 + 0.30*X4 + 0.40*X5 + 0.10*X6 + 0.10*X7 + 0.60*X8 + 0.80*X9 <= 0.40
prob +=  X1 + X2 + X3 + X4 + X5 + X6 + X7 + X8 + X9 == 1

prob.solve()
print(LpStatus[prob.status])
print("Min=", value(prob.objective))

# variables resultat
for v in prob.variables():
    print("%s=%.2f"%(v.name,v.varValue), end=', ')
````

Nous avons notre solution optimale&nbsp;! On peut même pousser le raffinement jusqu'à utiliser $\mathrm{\LaTeX}$ pour améliorer le rendu&nbsp;:

````python
from IPython.display import display, Latex

# variables resultat
result = ""
for v in prob.variables():
    valeur = round(v.varValue*100, 2)
    if valeur > 0:
        result += "$"+v.name+"="+str(valeur)+"\\%$, "

display(Latex(result))
````

*Une remarque&nbsp;: PuLP accepte aussi bien des problèmes de minimisation que de maximisation, et des contraintes d'infériorité que de supériorité. Mais tous les solveurs ne sont pas aussi souples, certains n'acceptent que programmes linéaires sous forme canonique pure, d'autres sous forme standard. Pourquoi&nbsp;? À votre avis, comment PuLP peut-il faire pour s'interfacer avec ces solveurs &nbsp;?*

PuLP doit faire du pré-traitement pour transformer le programme linéaire dans la forme canonique ou standard attendue par le solveur. Par exemple, pour les contraintes d'infériorité, PuLP peut introduire des variables d'écart pour les transformer en égalités. Pour les problèmes de maximisation, PuLP peut simplement multiplier la fonction objectif par -1 pour les transformer en problèmes de minimisation.

*Par ailleurs, l'algorithme du Simplexe n'est pas le seul sachant résoudre un programme linéaire. PuLP peut en utiliser un autre. Le connaissez-vous&nbsp;? Quels sont leurs principes généraux de fonctionnement, et leurs performances&nbsp;?*

L'algorithme du Simplexe est un algorithme de résolution de programmes linéaires qui fonctionne en parcourant les sommets de la région faisable pour trouver la solution optimale. Il est efficace pour les problèmes de petite à moyenne taille, mais peut être lent pour les problèmes de grande taille. PuLP peut également utiliser d'autres algorithmes tels que l'algorithme de points intérieurs, qui est plus efficace pour les problèmes de grande taille. L'algorithme de points intérieurs fonctionne en se déplaçant à l'intérieur de la région faisable plutôt que sur les sommets, ce qui peut conduire à une convergence plus rapide vers la solution optimale.

## 3.1 Modélisation du problème
*Comment pouvons-nous modéliser ce problème sous la forme d'un programme linéaire&nbsp;?*

En premier lieu, nous allons associer une variable binaire à chaque titre de la bibliothèque de notre étudiant. Cette variable prendra la valeur 1 si le titre est choisi pour être ajouté à la liste de lecture du smartphone, et 0 sinon. Nous allons ensuite définir une fonction objectif qui maximise la somme des notes des titres choisis, tout en respectant la contrainte de capacité du smartphone.

1. **Variables de décision** : Soit $X_i$ une variable binaire qui vaut 1 si le titre $i$ est choisi, et 0 sinon.
2. **Fonction objectif** : Maximiser la somme des notes des titres choisis, soit $\sum_{i=1}^{n} v_i X_i$, où $v_i$ est la note du titre $i$.
3. **Contrainte de capacité** : La somme des tailles des titres choisis ne doit pas dépasser la capacité du smartphone, soit $\sum_{i=1}^{n} w_i X_i \leq C$, où $w_i$ est la taille du titre $i$ et $C$ est la capacité du smartphone.
4. **Contraintes de variables** : $X_i \in \{0, 1\}$ pour tout $i$.
5. **Formulation du problème** :
\[
\begin{align*}
\text{Maximiser} \quad & \sum_{i=1}^{n} v_i X_i \\
\text{Sous les contraintes} \quad & \sum_{i=1}^{n} w_i X_i \leq C \\
& X_i \in \{0, 1\} \quad \text{pour tout } i
\end{align*}
\]


<br />

*Un algorithme naïf pour déterminer un résultat optimal serait une approche par force brute, qui teste toutes les solutions (admissibles ou non) et conserve la meilleure solution admissible. Dans le cas de notre étudiant, c'est une solution qui semble envisageable. Mais si l'étudiant supprime plusieurs Giga octets de jeux, et qu'il enrichit sa bibliothèque musicale, cette solution restera-t-elle viable&nbsp;? Qu'est-ce qui va devenir bloquant&nbsp;?*

Cette approche par force brute a une complexité exponentielle, car elle doit tester toutes les combinaisons possibles de titres. Si l'étudiant a n titres dans sa bibliothèque, il y a $2^n$ combinaisons possibles (car chaque titre peut être choisi ou non). Par conséquent, à mesure que le nombre de titres augmente, le temps nécessaire pour trouver la solution optimale devient rapidement impraticable. C'est ce qui rend cette approche non viable pour des bibliothèques musicales plus grandes.


## 3.2 Résolution

Puisque nous avons modélisé le problème sous forme linéaire, pourquoi ne pas tester l'algorithme du Simplexe dessus&nbsp;? Reprenons les données issues du Smartphone de notre étudiant. Attention aux bornes de nos variables&nbsp;!

````python
# variables du problème
X1  = LpVariable("X_01", 0, 1)
X2  = LpVariable("X_02", 0, 1)
X3  = LpVariable("X_03", 0, 1)
X4  = LpVariable("X_04", 0, 1)
X5  = LpVariable("X_05", 0, 1)
X6  = LpVariable("X_06", 0, 1)
X7  = LpVariable("X_07", 0, 1)
X8  = LpVariable("X_08", 0, 1)
X9  = LpVariable("X_09", 0, 1)
X10 = LpVariable("X_10", 0, 1)


# probleme
prob = LpProblem("Le_probleme_du_sac_a_dos", LpMaximize)

# objectif
prob += 2*X1 + 5*X2 + X3 + 2*X4 + 3*X5 + 5*X6 + 2*X7 + 3*X8 + X9 + X10, "Note totale"

# contrainte
prob += 40*X1 + 52*X2 + 27*X3 + 68*X4 + 80*X5 + 62*X6 + 63*X7 + 53*X8 + 56*X9 + 59*X10 <= 400, "Contrainte_de_poids"

prob.solve()
print(LpStatus[prob.status])
print("Valeur totale = ", value(prob.objective))

poids_total = 0.0
poids = [40, 52, 27, 68, 80, 62, 63, 53, 56, 59]
for idx, v in enumerate(prob.variables()):
    print("X%d=%d" % (idx+1, v.varValue), end=', ')
    poids_total += v.varValue*poids[idx]
print("\npoids=%d" % poids_total)
````

*Ça vous parait correct&nbsp;? Pourtant, cela ne l'est pas&nbsp;! Calculez à la main le poids de la solution, en reprenant les valeurs dans le tableau. Qu'obtenez-vous&nbsp;? Pourquoi&nbsp;?*

En calculant à la main, on trouve que le poids total de la solution est de 398, ce qui est inférieur à la capacité du smartphone (400). Cependant, la solution obtenue par le Simplexe n'est pas admissible car elle inclut une variable (X9) qui prend une valeur proche de 1, ce qui signifie que le titre correspondant est presque entièrement choisi. En réalité, dans un problème de sac à dos, les variables doivent être binaires (0 ou 1), et une variable qui prend une valeur fractionnaire n'est pas admissible. C'est pourquoi la solution obtenue par le Simplexe n'est pas valide pour ce problème spécifique.


Vérifions ça&nbsp;:

````python
poids_total=0.0
for idx, v in enumerate(prob.variables()):
    print("X%d=%f" % (idx+1, v.varValue), end=', ')
    poids_total += v.varValue*poids[idx]
print("\npoids=%d"%poids_total)
````

Et là ça ne va plus du tout, La solution n'est pas admissible. On a vu qu'un arrondi inférieur produisait une solution qui, elle, est admissible. Mais avec une valeur de $X_9$ aussi proche de 1, ce n'est peut-être pas la solution optimale. Est-ce qu'un <a href="https://www.w3schools.com/python/ref_func_round.asp">arrondi au plus proche</a> marcherait mieux&nbsp;?

````python
poids_total=0.0
for idx, v in enumerate(prob.variables()):
    print("X%d=%f" % (idx+1, v.varValue), end=', ')
    poids_total += int(round(v.varValue))*poids[idx]
print("\npoids=%d" % poids_total)
````

*Qu'observe-t-on&nbsp;?*

On observe que l'arrondi au plus proche a permis d'obtenir une solution admissible, mais elle n'est pas optimale. En effet, en arrondissant la variable X9 à 1, on inclut le titre correspondant dans la solution, ce qui augmente la valeur totale de la solution. Cependant, cela peut également entraîner une violation de la contrainte de capacité du smartphone, car le poids total peut dépasser la limite de 400. Par conséquent, l'arrondi au plus proche ne garantit pas une solution optimale pour ce problème de sac à dos.

*Parmi toutes les solutions générées, nous en avons une solution qui est admissible. Est-ce une bonne méthode pour trouver une solution optimale&nbsp;? Prenons cet exemple, dans lequel l'espace des solutions admissibles délimité par les deux contraintes est en blanc. Est-ce que, pour un problème _de maximisation_, arrondir une solution optimale réelle fonctionnerait&nbsp;?*


Dans un problème de maximisation, arrondir une solution optimale réelle peut ne pas fonctionner, car cela peut entraîner une violation de la contrainte de capacité du smartphone. En arrondissant une variable à 1, on peut inclure un titre qui dépasse la limite de poids, ce qui rend la solution inadmissible. Par conséquent, l'arrondi d'une solution optimale réelle ne garantit pas une solution admissible pour ce problème de sac à dos.

*C'est finalement assez logique, puisque le problème de décision est $\mathcal{NP}$-Complet ([on le sait depuis 1972](https://fr.wikipedia.org/wiki/21_probl%C3%A8mes_NP-complets_de_Karp)). Le nombre exponentiel de solution, qu'on a démontré tout à l'heure, était-il un indice&nbsp;?*

Oui, le nombre exponentiel de solutions était un indice que le problème de sac à dos est un problème difficile à résoudre de manière optimale. En effet, le fait qu'il y ait $2^n$ combinaisons possibles de titres à choisir indique que la complexité du problème augmente rapidement avec le nombre de titres dans la bibliothèque. Cela suggère que trouver une solution optimale peut être très difficile, voire impossible, pour des bibliothèques musicales plus grandes, ce qui est caractéristique des problèmes $\mathcal{NP}$-complets.
